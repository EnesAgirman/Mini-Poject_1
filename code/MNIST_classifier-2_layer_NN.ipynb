```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
import struct
import os
```

```python
mainFilePath = os.path.normpath(os.getcwd() + os.sep + os.pardir)

training_images_directory = r"data\training\images\train-images.idx3-ubyte"
training_labels_directory = r"data\training\labels\train-labels.idx1-ubyte"

test_images_directory = r"data\test\images\t10k-images.idx3-ubyte"
test_labels_directory = r"data\test\labels\t10k-labels.idx1-ubyte"

training_images_directory = os.path.join(mainFilePath, training_images_directory)
training_labels_directory = os.path.join(mainFilePath, training_labels_directory)
test_images_directory = os.path.join(mainFilePath, test_images_directory)
test_labels_directory = os.path.join(mainFilePath, test_labels_directory)


with open(training_images_directory, 'rb') as file:
        _, train_size, _, _ = struct.unpack(">IIII", file.read(16))

with open(test_images_directory, 'rb') as file:
        _, test_size, _, _ = struct.unpack(">IIII", file.read(16))

x_train = np.fromfile(training_images_directory, np.ubyte, offset=16).reshape(train_size, 784) / 255
y_train = np.fromfile(training_labels_directory, np.ubyte, offset=8)
x_test = np.fromfile(test_images_directory, np.ubyte, offset=16).reshape(test_size, 784) / 255
y_test = np.fromfile(test_labels_directory, np.ubyte, offset=8)
```

```python
plt.figure(figsize=(15, 150))
for i in range(0, 5):
    plt.subplot(1, 5, i+1)
    img = x_train[i].reshape(28, 28)
    plt.imshow(img, cmap=plt.cm.gray)
    plt.title(f"training image {i}, label = {y_train[i]}", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(15, 150))
for i in range(0, 5):
    plt.subplot(1, 5, i+1)
    img = x_test[i].reshape(28, 28)
    plt.imshow(img, cmap=plt.cm.gray)
    plt.title(f"training image {i}, label = {y_test[i]}", fontsize=10)
plt.show()
```
